## Inheritance --> 
        is the capability of one class to derive or inherit the properties from another class.
 - Parent class
    
 - child class

```python
#parent class
class Person:
    def __init__(self,name,idno):
        self.name=name
        self.idno=idno
    # instance method
    def display(self):
        print(self.name)
        print(self.idno)
    #instance method
    def details(self):
        print(self.name)
        print(self.idno)
            
#child class\
class Employee(Person):
    def __init__(self,name,idno,salary,post,age):
        self.salary=salary
        self.post=post
        self.age=age
        
        #invoke the __init__ of the parent class
        Person.__init__(self,name,idno)
        
    def display(self):
        print(self.name)
        print(self.idno)
        print(self.post)
        print(self.salary)
        
#create object variable (child class)
a=Employee('Ravi',88767,200000,"Intern",20)


# calling a function of the class Person using its instance
a.display()
a.details()        
```

##### Python Multi-Level inheritance
Multi-Level inheritance is possible in python like other object-oriented languages. Multi-level inheritance is archived when a derived class inherits another derived class. There is no limit on the number of levels up to which, the multi-level inheritance is archived in python.


```python
class Animal:    # parent class 
    
    def speak(self):  
        print("Animal Speaking")  
#The child class Dog inherits the base class Animal  
class Dog(Animal):  # child class
    def bark(self):  
        print("dog barking")  
#The child class Dogchild inherits another child class Dog  
class DogChild(Dog): #child class for Dog 
    def eat(self):  
        print("Eating bread...")  
d = DogChild()  
d.bark()  
d.speak()  
d.eat()  
```

##### Python Multiple inheritance
Python provides us the flexibility to inherit multiple base classes in the child class.


```python
class Calculation1: # parent class 
    def Summation(self,a,b):  
        return a+b;  
class Calculation2: #parent class 
    def Multiplication(self,a,b):  
        return a*b;  
class Derived(Calculation1,Calculation2): #child class 
    def Divide(self,a,b):  
        return a/b;  
d = Derived()  
print(d.Summation(10,20))  
print(d.Multiplication(10,20))  
print(d.Divide(10,20))  
```

#### Polymorphism --> Having many forms
Polymorphism in python defines methods in the child class that have the same name as the methods in the parent class

```python
class Bird:        # parent class
    def intro(self):
        print(" There are many types of birds")
    def flight(self):
        print("Most of the birds can fly but some cannot")
class sparrow(Bird):   # child class
    def flight(self):
        print("Sparrow can fly")
class ostrich(Bird):   # child class
    def flight(self):
        print("ostrich cannot fly")

obj_bird=Bird()
obj_spa=sparrow()
obj_ost=ostrich()


obj_bird.intro()
obj_bird.flight()

obj_spa.intro()
obj_spa.flight()

obj_ost.intro()
obj_ost.flight()
```

```python
class Square:
    side = 5     
    def calculate_area(self):
        return self.side * self.side

class Triangle:
    base = 5
    height = 4
    def calculate_area(self):
        return 0.5 * self.base * self.height

sq = Square()
tri = Triangle()
print("Area of square: ", sq.calculate_area())
print("Area of triangle: ", tri.calculate_area())
```

#### Overriding:
Method overriding in Python is when you have two methods with the same name that each perform different tasks. This is an important feature of inheritance in Python.

In method overriding, the child class can change its functions that are defined by its ancestral classes. 

There are two prerequisite conditions for Method overriding:

Inheritance should be present in the code, method overriding cannot be performed in the same class, and overriding can only be executed when a child class is derived through inheritance.

The child class should have the same name and the same number of parameters as the parent class.

```python
class Animal:
    def Walk(self):
        print('Hello, I am the parent class')

class Dog(Animal):
    def Walk(self):
        print('Hello, I am the child class')
        
print('The method Walk here is overridden in the code')

#Invoking Child class through object r

r = Dog()
r.Walk()

#Invoking Parent class through object r

r = Animal()
r.Walk()
```

The code enters the else block only if the try clause does not raise an exception

```python
# try, except and else
def divide(x,y):
    try:
        result=x//y
    except ZeroDivisionError:
        print('error:you are dividing by zero')
    else: 
        print(result)
divide(3,2)
divide(6,0)
```

##### Finally Keyword in Python
Python provides a keyword finally, which is always executed after the try and except blocks. The final block always executes after normal termination of try block or after try block terminates due to some exception

```python
# try,except, else and finally
def divide(x,y):
    try:
        result=x//y
    except ZeroDivisionError:
        print('you are dividing by zero')
    else: 
        print(result)
        
    finally:
        print('End')
divide(17,12)
```

##### Raising Exception
The raise statement allows the programmer to force a specific exception to occur. The sole argument in raise indicates the exception to be raised

```python
## raise and assert 
# raising exceptions in python
try:
    a=int(input())
    
    if a<=0:
        raise ValueError("that is not a positive number")
except ValueError as ve:
    print(ve)
```
